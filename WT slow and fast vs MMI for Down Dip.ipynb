```python
import UsefulFunctions as uf
import numpy as np
from matplotlib import pyplot as plt
```

```python
eq2 = uf.Earthquake('Data/Down Dip/grids/2_grid.xml')
eq4 = uf.Earthquake('Data/Down Dip/grids/4_grid.xml')
eq7 = uf.Earthquake('Data/Down Dip/grids/7_grid.xml')
eqs = [eq2,eq4,eq7]
```

```python
s=10
a=0.2
fs=14
ns = [2, 4, 7]
fig, ax = plt.subplots(1, 3, figsize=(8,5), sharey=True, sharex=True)
fig.subplots_adjust(wspace=0.005)

for i in range(3):
    if i == 2:
        label1 = 'Lower Range'
        label2 = 'Upper Range'
        ax[i].scatter(-1, -1, c='darkorange', alpha=1, s=s, label=label1, zorder=-2)
        ax[i].scatter(-1, -1, c='mediumblue', alpha=1, s=s, label=label2, zorder=-2)
        
    wtf = eqs[i].warning_times_s.copy()
    wts = eqs[i].warning_times_slow.copy()
    mmi = eqs[i].mmi.copy()
    
    ax[i].scatter(mmi, wtf, c='darkorange', alpha=a, s=s, zorder=4)
    ax[i].scatter(mmi, wts, c='mediumblue', alpha=a, s=s, zorder=4)
    ax[i].text(0.95, 0.97, s=f'Scenario {ns[i]}', ha='right', va='top', transform=ax[i].transAxes, fontsize=fs-2)
    # ax[i].yscale('log')
    ax[i].set_ylim(-20,120)
    ax[i].set_xlim(4.9, 8.33)
    ax[i].set_xticks(range(5,9))
    ax[i].grid(alpha=0.2)
    
ax[2].legend(loc=(0.29, .8), frameon=True, fancybox=True)
fig.text(0.05, 0.5, r'Warning Time $(s)$', va='center', rotation='vertical', fontsize=fs)
fig.text(0.5, 0.02, f'MMI', ha='center', fontsize=fs)
plt.savefig('Figures/Down Dip/3 full.png', dpi=700)
plt.show()
```
