```python
import os
import UsefulFunctions as uf
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
directory_in_str = 'Data/InterfaceMag/grids'
directory = os.fsencode(directory_in_str)

# kodiak lat lon
kodiak_lon = -152.41
kodiak_lat = 57.79

# eq = uf.Earthquake('Data/AncMag/grids/AncMag8_0.xml')
# eq_anc_index = uf.getNearestIndex(eq.lons, eq.lats, anc_lon, anc_lat)

# print(eq.warning_times_earlypeak[eq_anc_index])
# print(eq.warning_times_latepeak[eq_anc_index])
```

```python
counter = 0
epi_distances = np.array([10, 25, 50, 75, 100])
hypo_distances = (epi_distances**2 + 30.15811**2)**.5
data = pd.DataFrame(columns=['name', 'mag', 'duration', 'mmi', 'pga', 'wt_s', 'wt_early_peak', 'wt_late_peak'])
eq_dict = {}
distance_data_dict = {}
for dist in epi_distances:
    distance_data_dict[dist] = {}
    
for file in os.listdir(directory):
    filename = os.fsdecode(file)
#     print(f'''{file}
# {filename}''')
    if filename.endswith(".xml"):
        eq = uf.Earthquake(directory_in_str+'/'+filename)
        eq_kodiak_index = uf.getNearestIndex(eq.lons, eq.lats, kodiak_lon, kodiak_lat)
        eq_name = eq.event['event_id']
        eq_mag = eq.event['magnitude']
        eq_dur = eq.duration
        eq_mmi = eq.mmi[eq_kodiak_index][0]
        eq_pga = eq.pga[eq_kodiak_index][0]
        eq_wt_s = eq.warning_times_s[eq_kodiak_index][0]
        eq_wt_ep = eq.warning_times_earlypeak[eq_kodiak_index][0]
        eq_wt_lp = eq.warning_times_latepeak[eq_kodiak_index][0]
        data = data.append({
            'name': eq_name,
            'mag': eq_mag,
            'duration': eq_dur,
            'mmi': eq_mmi,
            'pga': eq_pga,
            'wt_s': eq_wt_s,
            'wt_early_peak': eq_wt_ep,
            'wt_late_peak': eq_wt_lp},
            ignore_index=True)
        
        eq_dict[eq_mag] = eq
        
        
        for i in range(epi_distances.size):
            dist = epi_distances[i]
            dist_indices = np.where(np.logical_and(eq.distances_epi.flatten()>=dist-5, eq.distances_epi.flatten()<=dist+5))
            distance_data_dict[dist][eq_mag] = {'mmi': eq.mmi.flatten()[dist_indices], 
                                                'pga': eq.pga.flatten()[dist_indices],
                                                'lons': eq.lons.flatten()[dist_indices],
                                                'lats': eq.lats.flatten()[dist_indices]}
        
        counter +=1
        print(f'Done with {counter} grids')
    else:
        continue
        
```

```python
data_sorted = data.sort_values(by=['mag'])
data_sorted.to_csv('Data/InterfaceMag/InterfaceMag_data.csv')
```

```python
dist = 100
gm = 'mmi'
for dist in epi_distances:
    plt.figure()

    for mag in distance_data_dict[dist]:
        mag_arr = mag*np.ones(distance_data_dict[dist][mag]['mmi'].size)
        plt.scatter(mag_arr, distance_data_dict[dist][mag][gm], c='k', alpha=0.005)
        if gm == 'pga':
            plt.yscale('log')
        plt.grid(alpha=0.2, which='both')
        plt.xlabel('magnitude')
        plt.ylabel(f'ground motion ({gm})')
        plt.title(f'{dist} km')
```

```python

plt.rc('axes', axisbelow=True)
colors = ['purple', 'red', 'orange', 'gold', 'green', 'blue']
ci = 0
s = 10
gm = 'mmi'
plt.figure()
for i in range(len(epi_distances)):
    dist = epi_distances[i]
    hdist = np.around(hypo_distances[i], 1)
    c = colors[i]
    for mag, dat in distance_data_dict[dist].items():
        if mag == 5:
            plt.scatter(mag, dat['mmi'].mean(), c=c, s=s, label=f'{dist} km ({hdist} km hypo)')
        else:
            plt.scatter(mag, dat['mmi'].mean(), c=c, s=s)
    
    ci += 1
if gm == 'pga':
    plt.yscale('log')
plt.xlabel('magnitude')
plt.ylabel(f'ground motion ({gm})')
plt.grid(alpha=0.2)

plt.legend(loc='lower right')
plt.show()
```
